```python
# Importando las bibliotecas necesarias para visualización de datos y análisis exploratorio
import matplotlib.pyplot as plt  # Biblioteca para crear gráficos y visualizaciones estáticas, animadas e interactivas en Python
import seaborn as sns  # Biblioteca para crear gráficos estadísticos atractivos y informativos en Python
import pandas as pd #Biblioteca la cual es una poderosa herramienta para manipulación y análisis de datos.
```

```python
# Leer datos desde archivo CSV
csv_path="https://raw.githubusercontent.com/LauraVas1/ProyectoF/main/diabetes.csv"
laura = pd.read_csv(csv_path, sep=",")

# Crear la lista headers
headers = ["Embarazos","Glucosa","Presion-arterial","Grosor-piel","Insulina","IMC","Funcion-PedigriDiabetes","Edad","Resultado"] #IMC=ÍNDICE DE MASA CORPORAL
laura.columns = headers

laura.head()
```

```python
# La función dtype genera una tabla con el tipo de dato de cada columna
laura.dtypes
```

```python
#Reemplazar 1 por "Positivo" y 0 por "Negativo"
laura['Resultado'] = laura['Resultado'].replace({1: 'Positivo', 0: 'Negativo'})
laura.head()
```

```python
#Comprobación de la cantidad de filas y columnas del conjunto de datos
laura.shape
```

```python
#count: Número total de observaciones o elementos en un conjunto de datos.
#mean: El valor promedio de un conjunto de datos.
#std: Mide la dispersión o la variabilidad de un conjunto de datos.
#min: El valor más pequeño en un conjunto de datos.
#max: El valor más grande en un conjunto de datos.
laura.describe()
```

```python
# Proporciona información concisa sobre un DataFrame llamado "laura".
laura.info()
```

```python
# La función dtype genera una tabla con el tipo de dato de cada columna.
laura.dtypes
```

```python
#Eliminar filas con valores nulos
laura= laura.dropna()

# Eliminar filas duplicadas basadas en todas las columnas
laura = laura.drop_duplicates()

laura.head()
```

```python
# Creando un histograma para la columna 'Edad' del DataFrame laura
# figsize=(5,5) especifica las dimensiones del gráfico en pulgadas
laura['Edad'].hist(figsize=(5,5))

# Mostrando el gráfico en pantalla
plt.show()
```

```python
# Utilizando Seaborn para crear un diagrama de distribución (histograma y KDE) para la columna 'Edad' del DataFrame laura
sns.distplot(laura['Edad'])
```

```python
# Creando una figura con dimensiones específicas (10x10 pulgadas)
fig=plt.figure(figsize=(5,5))

# Creando un diagrama de caja (boxplot) para la columna 'price' del DataFrame df
# Esto ayuda a visualizar la distribución de los datos, identificar outliers, y observar la tendencia central y dispersión
sns.boxplot(y=laura["Embarazos"])

# Mostrando el gráfico en pantalla
plt.show()
```

```python
# Calculando estadísticas descriptivas para la columna 'Edad' del DataFrame laura
mean = laura['Embarazos'].mean()  # Calcula la media (promedio) de 'engine-size'
median = laura['Embarazos'].median()  # Calcula la mediana de 'engine-size'
mode = laura['Embarazos'].mode()  # Calcula la moda de 'engine-size'
skew = laura['Embarazos'].skew()  # Calcula la asimetría (skewness) de 'engine-size'
kurt = laura['Embarazos'].kurt()  # Calcula la curtosis (kurtosis) de 'engine-size'

print("La media es :",mean)
print("La mediana es :",median)
print("La moda es :",mode)
print("El sesgo es :",skew)
print("La kurtosis es :",kurt)
```

```python
# Creando un histograma para la columna 'Resultado' del DataFrame laura
# figsize=(5,5) especifica las dimensiones del gráfico en pulgadas
laura['Resultado'].hist(figsize=(5,5))

# Mostrando el gráfico en pantalla
plt.show()
```
